```python
import random
import time
```

```python
class node(object):
    def __init__(self, number = None):
        self.number = number
        self.left = None
        self.right = None
        self.parent = None
```

```python
class tree(object):
    def __init__(self):
        self.dummy = node()
    
    def Add(self, num, nd = None):
        if(nd == None):
            nd = self.dummy.right
        if(nd == None):    # jeśli drzewo było puste
            self.dummy.right  = node(num)
            self.dummy.right.parent = self.dummy
            return
        if(nd.number>num):
            if(nd.left == None):
                nd.left = node(num)
                nd.left.parent = nd
                return
            else:
                self.Add(num, nd.left)
                return
        elif(nd.number<num):
            if(nd.right == None):
                nd.right = node(num)
                nd.right.parent = nd
                return
            else:
                self.Add( num, nd.right)
                return
        else:
            return

        
    def Find(self, num, nd = "u"):
        # funkcja ma zwrocić węzeł o zadanym kluczu, jeśli taki w drzewie jest
        # i zwrócić None, jeśli go nie ma.
        if(nd == "u"):
            nd = self.dummy.right
        if(nd == None):    # jeśli drzewo było puste
            return None
        if(nd.number==num):
            return nd
        if(nd.number>num):
            return self.Find(num, nd.left)
        else:
            return self.Find(num, nd.right)
        
        
        

    def Myk(self, nd = "u", depth = 0):
        if(nd == "u"):
            nd = self.dummy.right
        if (nd == None):
            return
        if nd.right != None:
            self.Myk(nd.right, depth + 1)
        for _ in range(0, depth):
            print("   ",end = "")
        print(nd.number)
        if nd.left != None:
            self.Myk(nd.left, depth + 1)

    def MinOfTree(self, nd = "u"):
        if nd == "u":
            nd = self.dummy.right
        current = nd
        while(current.left is not None):
            current = current.left
        return current
    
    def DeleteNode(self, num):
        nd = Find(self, num)
        if nd == None:
            return
        if nd.right is None and nd.left is None:
            if nd.parent.left == nd:
                nd.parent.left = None
            else:
                nd.parent.right = None
            return    
        
        if nd.right == None or nd.left == None:
            #TOC -> the only child
            TOC = nd.left
            if TOC == None:
                TOC = nd.right
            if nd.parent.left == nd:
                nd.parent.left = TOC
            else:
                nd.parent.right = TOC
            TOC.parent = nd.parent
            return
        
        NextNode = MinOfTree(nd.right)
        nd.number == NextNode.number
        DeleteNode(NextNode)
        return
    
    def Rotate(self, num):
        B = Find(self, num)
        if B is None or B == self.dummy or B == self.dummy.right:
            return
        A = B.parent
        P = A.parent
        if A.right == B: #left rotate
            Beta = B.left
            B.parent = P
            if P.left == A:
                P.left = B
            else:
                P.right = B
            B.left = A
            A.parent = B
            A.right = Beta
            if Beta != None:
                Beta.parent = A
            return
        else:
            Beta = B.right
            B.parent = P
            if P.right == A:
                P.right = B
            else:
                P.left = B
            B.right = A
            A.parent = B
            A.left = Beta
            if Beta != None:
                Beta.parent = A
            return
        
```

```python
t = tree()

how_much = 10



start_time = time.time()




for _ in range(how_much):
    x = random.randint(1,50)
    print(x, end = "-")
    t.Add(x)

print("\n\n\n ")


 
t.Myk()



"""
print(" Podaj liczbę ")

szukana = input()

n = t.Find(int(szukana))
if n==None:
    print("nie znalazłem")
else:
    print("znalazłem")
    print("jego ojcem jest ",n.parent.number)
    print(" ")
    t.Myk()
    
elapsed_time = time.time()- start_time
#print(elapsed_time)
"""
```
